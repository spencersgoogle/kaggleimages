```python
!uv pip install gradio
# !uv pip install gradio==5.31
import gradio
import pandas as pd
import plotly.graph_objects as go

csv_url_ibm = "https://raw.githubusercontent.com/MainakRepositor/Datasets/refs/heads/master/Stocks/IBM.csv"
df_ibm = pd.read_csv(csv_url_ibm)

csv_url_amzn = "https://raw.githubusercontent.com/MainakRepositor/Datasets/refs/heads/master/Stocks/AMZN.csv"
df_amzn = pd.read_csv(csv_url_amzn)

# IBM Chart
fig_ibm = go.Figure(data=[go.Candlestick(x=df_ibm['Date'],
                                         open=df_ibm['Open'],
                                         high=df_ibm['High'],
                                         low=df_ibm['Low'],
                                         close=df_ibm['Close(t)'])])
fig_ibm.update_layout(title='IBM Stock Chart', xaxis_title='Date', yaxis_title='Price')

# Amazon Chart
fig_amzn = go.Figure(data=[go.Candlestick(x=df_amzn['Date'],
                                          open=df_amzn['Open'],
                                          high=df_amzn['High'],
                                          low=df_amzn['Low'],
                                          close=df_amzn['Close(t)'])])
fig_amzn.update_layout(title='Amazon Stock Chart', xaxis_title='Date', yaxis_title='Price')


# Create Gradio App using Blocks not interfaces
# Blocks gives you more control over layout than interfaces
with gradio.Blocks() as demo:
    gradio.Markdown("## Stock Analysis")


    with gradio.Row():
      # side by side with gradio.Row:
      with gradio.Tabs():
        #tabs with gradio.Tabs
        with gradio.TabItem("IBM Chart"):
            gradio.Plot(value=fig_ibm)
        with gradio.TabItem("IBM Data"):
            gradio.DataFrame(value=df_ibm.head(20))

      with gradio.Tabs():
        with gradio.TabItem("AMZN Chart"):
            gradio.Plot(value=fig_amzn)
        with gradio.TabItem("AMZN Data"):
            gradio.DataFrame(value=df_amzn.head(20))

    # Standalone Tabs:
    with gradio.Tabs():
      with gradio.TabItem("IBM"):
            gradio.Plot(value=fig_ibm)
      with gradio.TabItem("Amazon"):
            gradio.Plot(value=fig_amzn)

    # Standalone Chart
    gradio.Plot(value=fig_ibm)
    gradio.Markdown("### IBM Raw Data")
    # Display the IBM DataFrame using gradio.DataFrame
    gradio.DataFrame(value=df_ibm.head(20))

    # Display the second chart
    gradio.Plot(value=fig_amzn)
    gradio.Markdown("### Amazon Raw Data")
    # Display the IBM DataFrame using gradio.DataFrame
    gradio.DataFrame(value=df_amzn.head(20))

# This will start the Gradio server and print an access URL
# demo.launch()

import pandas as pd

# Create a sample DataFrame
data = {'col1': [1, 2, 3], 'col2': ['A', 'B', 'C']}
df = pd.DataFrame(data)
```
