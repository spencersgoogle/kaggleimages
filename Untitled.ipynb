```python
import pandas as pd
import numpy as np

# Create a sample DataFrame
data = {
    'Category': ['A', 'B', 'C', 'D', 'E', 'A', 'B', 'C', 'D', 'E'],
    'Value1': np.random.randint(10, 100, 10),
    'Value2': np.random.randint(50, 150, 10),
    'Date': pd.to_datetime(pd.date_range(start='2023-01-01', periods=10, freq='D'))
}
df = pd.DataFrame(data)

print("Sample DataFrame created:")
```

```python
# Display the first 5 rows of the DataFrame
display(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a bar plot of 'Value1' by 'Category'
plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Value1', data=df)
plt.title('Bar Plot of Value1 by Category')
plt.xlabel('Category')
plt.ylabel('Value 1')
plt.show()
```

```python
# Create a scatter plot of 'Value1' vs 'Value2'
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Value1', y='Value2', hue='Category', data=df, s=100)
plt.title('Scatter Plot of Value1 vs Value2')
plt.xlabel('Value 1')
plt.ylabel('Value 2')
plt.grid(True)
plt.show()
```
